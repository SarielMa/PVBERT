```python
def my_clean(txt):
    txt = txt.strip()
    txt = " ".join(txt.split())
    return txt

def remove_pinfo(data):
    ret = []
    author_set = ["From Patient:", "From Provider:"]
    for item in data:
        #mid = item["messageID"]
        txt = my_clean(item["text"])
        for a in author_set:
            if a in txt:
                new_txt = my_clean(txt.split(a)[1])
        #author = transform[mid]
        #new_txt = f"{author}: {txt}"
        ret.append({"text": new_txt, "labels": item["labels"]})
    return ret
new_data1 = remove_pinfo(data1)
new_data2 = remove_pinfo(data2) 
```
