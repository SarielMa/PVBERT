```python
import json
import shutil
from pathlib import Path

# Path to your output_dir
output_dir = Path("/home/lm2445/palmer_scratch/results_071325_class_topic_old")
target_dir = Path("/home/lm2445/palmer_scratch/results_071325_class_topic")
target_dir.mkdir(exist_ok=True)

models = [
    "dmis-lab/biobert-v1.1",
    "allenai/scibert_scivocab_uncased",
    "emilyalsentzer/Bio_ClinicalBERT",
    "cambridgeltl/SapBERT-from-PubMedBERT-fulltext",
    "bert-base-uncased",
    "bert-large-uncased",
    "Twitter/twhin-bert-base",
    "eppc_bert_base",
    "eppc_bert_large"
]

# Metric to use
metric_name = "eval_micro_f1"
prefix = "eppc_model_"
# Loop through each model folder
for model_name in models:
    model_name = prefix + model_name
    model_dir = output_dir / model_name  # Now a Path
    if not model_dir.is_dir():
        print(f"⚠️ Skipping: {model_dir} is not a directory.")
        continue

    print(f"\n🔍 Processing model: {model_name}")
    
    # Find all checkpoints under this model folder
    checkpoints = sorted(model_dir.glob("checkpoint-*"))
    scores = []

    for ckpt in checkpoints:
        trainer_state_file = ckpt / "trainer_state.json"
        if trainer_state_file.exists():
            with open(trainer_state_file) as f:
                state = json.load(f)
            
            log_history = state.get("log_history", [])
            eval_results = [entry for entry in log_history if "eval_loss" in entry]
            
            if eval_results:
                last_eval = eval_results[-1]
                score = last_eval.get(metric_name, None)
                if score is not None:
                    scores.append((ckpt, score))
    
    if not scores:
        print(f"⚠️ No valid checkpoints found for {model_name}")
        continue

    # Sort by metric: higher is better for f1
    scores.sort(key=lambda x: x[1], reverse=True)

    # Pick best 3
    best_3 = scores[:3]

    # Copy each to target_dir/model_name/checkpoint-*
    for ckpt, score in best_3:
        dest = target_dir / model_name / ckpt.name
        dest.parent.mkdir(parents=True, exist_ok=True)
        shutil.copytree(ckpt, dest)
        print(f"✅ Copied {ckpt.name} with {metric_name}={score:.4f}")

print(f"\n🎉 All best checkpoints saved under: {target_dir}")
```
