```python
file_path1 = "stratified_test_data_old.json"  # Update with your actual file path
with open(file_path1, "r") as f:
    data1 = json.load(f)
file_path2 = "stratified_train_data_old.json"  # Update with your actual file path
with open(file_path2, "r") as f:
    data2 = json.load(f)
```

```python
# train_data = []
# test_data = []

to_remove = ["InfoSeek_carecoordination"]
to_remove = set(to_remove)
def remove_labels(dataset, to_remove = to_remove):
    test_data = []
    count = 0
    for data in dataset:
        labels = data["labels"]
        new_labels = []
        for label in labels:
            if label not in to_remove:
                new_labels.append(label)
            else:
                count += 1
        assert type(new_labels) == type(labels)
        if len(new_labels) != 0:
            test_data.append({"text":  data["text"], "labels": new_labels})
    print (f" lenght after removing is {len(test_data)}")
    print (f"number of labels removed {count}")
    return test_data
    
test_data = remove_labels(data1)
train_data = remove_labels(data2)
```

```python
with open("stratified_train_data.json", "w") as f:
    json.dump(train_data, f, indent=4)

with open("stratified_test_data.json", "w") as f:
    json.dump(test_data, f, indent=4)
```
